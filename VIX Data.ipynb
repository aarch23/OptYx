```python
import pandas as pd
import numpy as np 
from pathlib import Path 
```

```python
# convert vix 30 min Jan data from a txt file into a csv

vix_jan_raw = pd.read_csv("vix_30x_202201.txt")
#vix_jan_raw.to_csv("vix_30x_202201.csv", index=None)
```

```python
vix_feb_raw = pd.read_csv("vix_30x_202202.txt")
#vix_feb_raw.to_csv("vix_30x_202202.csv", index=None)
```

```python
vix_mar_raw = pd.read_csv("vix_30x_202203.txt")
#vix_mar_raw.to_csv("vix_30x_202203.csv", index=None)
```

```python
vix_raw_data = pd.concat([vix_jan_raw, vix_feb_raw, vix_mar_raw])
```

```python
vix_raw_data
```

```python
# drop the puts columns

vix_raw_data.drop(vix_raw_data.iloc[:, 20:31], inplace = True , axis=1)

vix_raw_data.head()
```

```python
vix_raw_data.columns
```

```python
vix_raw_data.drop(vix_raw_data.columns.difference([' [UNDERLYING_LAST]', ' [QUOTE_READTIME]']), 1, inplace=True)
```

```python
vix_raw_data.drop_duplicates()
```

```python
vix_raw_data.to_csv("vixprice.csv", index=None) 
```
