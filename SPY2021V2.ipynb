# __START HERE to calculate 'y' and filter DTE__

```python
import pandas as pd 
import numpy as np 
from pathlib import Path
```

```python
spy_path = Path('spy_2021callsv1.csv')
spy_21_q4 = pd.read_csv(spy_path)
```

```python
spy_21_q4 = spy_21_q4.set_index('CONTRACT')
```

```python
spy_21_q4["PRICECLOSE"] = spy_21_q4.loc[spy_21_q4['DTE'] == 2.0, 'C_LAST']
spy_21_q4.head()
```

```python
spy_21_q4['ROI3'] = spy_21_q4["PRICECLOSE"] / spy_21_q4['C_LAST']
spy_21_q4.head()
```

```python
spy_21_q4['ROI %'] = (spy_21_q4['ROI3'] - 1) * 100
spy_21_q4.head(100)
```

```python
spy_21_q4 = spy_21_q4.drop(columns=['ROI3'])
spy_21_q4.head()
```

```python
#APPROACH A: 6 categories

#spy_21_q4.loc[spy_21_q4["ROI %"] <= -60, "y"] = 0
#spy_21_q4.loc[(spy_21_q4["ROI %"] <= -10) & (spy_21_q4["ROI %"] > -60), "y"] = 1
#spy_21_q4.loc[(spy_21_q4["ROI %"] <= 10) & (spy_21_q4["ROI %"] > -10), "y"] = 2
#spy_21_q4.loc[(spy_21_q4["ROI %"] <= 60) & (spy_21_q4["ROI %"] > 10), "y"] = 3
#spy_21_q4.loc[(spy_21_q4["ROI %"] > 60) & (spy_21_q4["ROI %"] < 100), "y"] = 4
#spy_21_q4.loc[spy_21_q4["ROI %"] >= 100, "y"] = 5
```

```python
#APPROACH B: 3 categories 

spy_21_q4.loc[spy_21_q4["ROI %"] <= -30, "y"] = 0

spy_21_q4.loc[(spy_21_q4["ROI %"] <= 25) & (spy_21_q4["ROI %"] > -30), "y"] = 1

spy_21_q4.loc[spy_21_q4["ROI %"] > 25, "y"] = 2
```

```python
spy_21_q4["y"].value_counts()
```

```python
spy_21_q4.rename(columns = {'UNDERLYING_LAST': 'SPY PRICE'}, inplace = True)
spy_21_q4.head()
```

```python
spy_21_q4.reset_index(inplace=True)
```

```python
spy_21_q4 = spy_21_q4.set_index('QUOTE_READTIME')
spy_21_q4.head()
```

```python
spy_21_q4.index = pd.to_datetime(spy_21_q4.index, infer_datetime_format=True)
spy_21_q4.head()
```

```python
#Drops all rows after DTE2.0

spy_21_q4_filt = spy_21_q4.loc[(spy_21_q4["DTE"]<=4.27) & (spy_21_q4["DTE"]>=2.0)]
```

```python
spy_21_q4_filt.shape
```

```python
spy_21_q4_filt = spy_21_q4_filt.dropna()
```

```python
spy_21_q4_filt.shape
```

```python
spy_21_q4_filt.to_csv("./Resources/spy_21_q4_filt.csv", index=True)
```
