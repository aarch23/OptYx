## Y Classification

# Create a 'y' column for our % profit/loss categories (0,1,2,3,4) 
# in % terms
# 0 is <-60 = strong sell
# 1 is (>-60 to -10) = sell
# 2 is (>-10 to 10) = pass
# 3 is (>10-60) = buy
# 4 is >60 to <80 = strong buy
# 5 is <100 = high return


```python
spy_data_v7.loc[spy_data_v7["ROI %"] <= -60, "y"] = 0
spy_data_v7.loc[(spy_data_v7["ROI %"] <= -10) & (spy_data_v7["ROI %"] > -60), "y"] = 1
spy_data_v7.loc[(spy_data_v7["ROI %"] <= 10) & (spy_data_v7["ROI %"] > -10), "y"] = 2
spy_data_v7.loc[(spy_data_v7["ROI %"] <= 60) & (spy_data_v7["ROI %"] > 10), "y"] = 3
#spy_data_v7.loc[spy_data_v7["ROI %"] > 60, "y"] = 4
spy_data_v7.loc[(spy_data_v7["ROI %"] > 60) & (spy_data_v7["ROI %"] < 100), "y"] = 4
spy_data_v7.loc[spy_data_v7["ROI %"] >= 100, "y"] = 5
```

```python
#Inflation Data 
# Nov - 0.8% - For dec-jan
# Dec +0.5% - For jan-feb
# Jan +0.6% - For feb-march
# Feb +0.8% - for March-april
```
